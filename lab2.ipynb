```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
def lin_random_gen(module, a, c, x_0, n):
  result = [x_0]
  for i in range(1, n+1):
    result.append((result[i-1] * a + c) % module)
  return result[1:]


# def find_period(n, X):
#     k = n + 1
#     period = 0
#     value = X[k]
#     find = X[n]

#     while find != value:
#         k += 1
#         value = X[k]
#         period += 1

#     return period


def find_period(sequence):
    n = len(sequence)
    period = 0
    for i in range(1, n):
        if (sequence[:i] == sequence[i:2*i]).all():
            period = i
            break
    return period

```

```python
r = lin_random_gen(1733, 1721, 1723, 26, 10000)
# plt.scatter(np.arange(0, len(r), 1), r)
plt.hist(r,alpha=0.5,bins=50, color="green")
# plt.ylim([0, 800])
plt.show()
```

```python
find_period(np.array(r))
```

```python
r10 = np.mod(r, 10)
print(find_period(r10))
plt.hist(r10,bins=10,alpha=0.5, color="green")
plt.show()
```

```python
r100 = np.mod(r, 100)
print(find_period(r100))
plt.hist(r100,bins=100,alpha=0.5, color="green")
plt.show()
```

```python
k = 0
newX = []
Y = []
S = []
for i in range(7000):
  s = 10
  while s == 0 or s > 1:
    k = k + 1
    x_generated = 2 * r[k] / 1733 - 1
    y_generated = 2 * r[k + 1] / 1733 - 1
    s = x_generated ** 2 + y_generated ** 2
  newX.append(x_generated)
  Y.append(y_generated)
  S.append(s)
```

```python
z1 = newX * np.sqrt(-2 * np.log(S) / S)
z2 = Y * np.sqrt(-2 * np.log(S) / S)
```

```python
plt.hist(z1, alpha=0.5, bins=20, color="green", density=True)
xx = np.arange(-3, 3, 0.01)
sigma, mu = 1, 0
plt.plot(xx, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (xx - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
plt.show()
```
